```python
# Calculate the correlation matrix
churn_corr = churn_df.corr()

# Set the threshold for high correlation
threshold = 0.8

# Set the lower triangle of the correlation matrix to zero
churn_corr.loc[:,:] = np.tril(churn_corr, k=-1)

# Stack the correlation matrix to a Series
churn_corr_series = churn_corr.stack()

# Find the highly correlated pairs
high_corr_pairs = churn_corr_series[(churn_corr_series > threshold) | (churn_corr_series < -threshold)]

# Get the list of columns to drop
cols_to_drop = set()
for pair in high_corr_pairs.index:
    cols_to_drop.add(pair[1])

# Drop the highly correlated columns
churn_df = churn_df.drop(columns=cols_to_drop)

# Print the remaining columns
print(churn_df.columns)
```

```python
#Here Applying SMOTE
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state = 42)
X_res,y_res = sm.fit_resample(X,y)
```

```python
#Model-1 Building
```

```python
#spiliting the data set into X_train, y_train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_res,y_res,train_size=0.7,random_state=25)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
from sklearn.metrics import accuracy_score, roc_curve, confusion_matrix, classification_report
```

```python
# Perform feature selection using RFE
model = LogisticRegression()
rfe = RFE(model, n_features_to_select=10)
rfe.fit(X_train, y_train)
selected_features = X_train.columns[rfe.support_]
```

```python
# Create new dataset with selected features
X_train_new = X_train[selected_features]
X_test_new = X_test[selected_features]

# Train the logistic regression model
lr = LogisticRegression()
lr.fit(X_train_new, y_train)
```

```python
# Make predictions on the testing set
y_pred = lr.predict(X_test_new)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
#ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
plt.plot(fpr, tpr)
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
```

```python
confusion_matrix = confusion_matrix(y_test, y_pred)
print("Confusion matrix:\n", confusion_matrix)

print(classification_report(y_test, y_pred))
```

```python
#Model-2 Build
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
# Define the hyperparameters to tune
param_grid = {'max_depth': [3, 4, 5, 6],
              'min_samples_split': [2, 4, 6, 8],
              'min_samples_leaf': [1, 2, 3, 4]}
```

```python
# Create a decision tree classifier object
dt = DecisionTreeClassifier()
```

```python
# Create a GridSearchCV object to tune the hyperparameters
grid_search = GridSearchCV(dt, param_grid, cv=5, scoring='accuracy')
```

```python
# Fit the GridSearchCV object to the training data
grid_search.fit(X_train, y_train)
```

```python
# Get the best hyperparameters
best_params = grid_search.best_params_
```

```python
# Train a decision tree model with the best hyperparameters
dt = DecisionTreeClassifier(max_depth=best_params['max_depth'], min_samples_split=best_params['min_samples_split'], min_samples_leaf=best_params['min_samples_leaf'])
dt.fit(X_train, y_train)
```

```python
# Make predictions on the testing set
y_pred = dt.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
print(classification_report(y_test, y_pred))
```
